WEB SCRAPPING CON SELENIUM:

WEBSITE: quotes to scrape

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd

chrome_service = ChromeService(ChromeDriverManager().install())

options = webdriver.ChromeOptions()
options.add_argument("--headless")  

driver = webdriver.Chrome(service=chrome_service, options=options)

# Vamos a crear una lista de DataFrames
df_lista = []

# accedemos a la página web
url = "http://quotes.toscrape.com"

while url:
    driver.get(url)

    # Extraemos los datos de citas, autores y etiquetas
    quote_elements = driver.find_elements(By.CSS_SELECTOR, ".quote")
    for quote_element in quote_elements:
        quote_text = quote_element.find_element(By.CSS_SELECTOR, ".text").text
        author = quote_element.find_element(By.CSS_SELECTOR, ".author").text
        tags = [tag.text for tag in quote_element.find_elements(By.CSS_SELECTOR, ".tag")]

        # Creamos un DataFrame para cada cita
        df_quotes1 = pd.DataFrame({
            'Cita': [quote_text],
            'Autor': [author],
            'Etiquetas': [tags]
        })

        # Agregamos el DataFrame a la lista
        df_lista.append(df_quotes1)

    # Para hacer algo diferente buscaremos las citas de todas las páginas, no solo de la primera como habíamos hacho anteriormente con BeautifulSoup
    next_page = driver.find_elements(By.CSS_SELECTOR, ".next a")
    url = next_page[0].get_attribute("href") if next_page else None

# Concatenamos todos los DataFrames en uno solo
df_quotes_sel = pd.concat(df_lista, ignore_index=True)

print(df_quotes_sel)

# Cerramos el navegador
driver.quit()

```

Tria una pàgina web que tu vulguis i realitza web scraping mitjançant la llibreria Selenium primer i Scrapy després. 

Haremos web scraping de la lista de 100 películas más populares de IMDB. Primero con selenium y depués con Scrapy.

Con Selenium:

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
import pandas as pd

# Configuramos el driver de Selenium para Chrome
driver = webdriver.Chrome(service=Service(ChromeDriverManager().install()))

# Vamos a la página de IMDB
driver.get('https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm')

# Buscamos todas las filas 
rows = driver.find_elements(By.XPATH, '//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul/li')

# Creamos listas vacías donde almacenaremos la información de las películas
ratings = []
titles = []
years = []
durations = []

# Utilizamos Xpath para extraer la información de las filas
for row in rows:
    try:
        rating = row.find_element(By.XPATH, './div[2]/div/div/span/div/span').text
    except NoSuchElementException:
        rating = 'N/A'
    try:
        title = row.find_element(By.XPATH, './div[2]/div/div/div[2]/a/h3').text
    except NoSuchElementException:
        title = 'N/A'
    try:
        year = row.find_element(By.XPATH, './div[2]/div/div/div[3]/span[1]').text
    except NoSuchElementException:
        year = 'N/A'
    try:
        duration = row.find_element(By.XPATH, './div[2]/div/div/div[3]/span[2]').text 
    except NoSuchElementException:
        duration = 'N/A'

    # Añadimos la información a las listas
    ratings.append(rating)
    titles.append(title)
    years.append(year)
    durations.append(duration)

# Cerramos el navegador
driver.quit()

# Creamos un DataFrame 
df = pd.DataFrame({
    'Título': titles,
    'Año': years,
    'Duración': durations,
    'Puntuación IMDb': ratings
})


df
```

Web Scraping con Scrapy:

Ahora haremos web scraping de la misma lista de películas de IMDB con Scrapy.

Trabajaremos básicamente desde el terminal.

Primero crearemos un proyecto de scrapy:  'scrapy startproject imdb_scraper'

Una vez creado, dentro del directorio, encontraremos la carpeta spiders y ahí crearemos la spider, en nuestro caso con el nombre 'imdb_spider.py'. 

En la spider (anexaré copia de la misma al repositotio Github) copiaremos el siguiente código (no lo ejecutaremos en jupyter, por eso lo dejo como markdown a efectos ilustrativos):


import scrapy

class IMDBSpider(scrapy.Spider):
    name = "imdb_spider"
    start_urls = ['https://www.imdb.com/chart/moviemeter/?ref_=nv_mv_mpm']
    custom_settings = {
        'USER_AGENT': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
    }    
    
    def parse(self, response):
        for movie in response.xpath('//*[@id="__next"]/main/div/div[3]/section/div/div[2]/div/ul/li'):
            try:
                rating = movie.xpath('./div[2]/div/div/span/div/span/text()').get()
                title = movie.xpath('./div[2]/div/div/div[2]/a/h3/text()').get()
                year = movie.xpath('./div[2]/div/div/div[3]/span[1]/text()').get()
                duration = movie.xpath('./div[2]/div/div/div[3]/span[2]/text()').get()
            except Exception:
                continue

            yield {
                'Título': title,
                'Año': year,
                'Duración': duration,
                'Rating': rating
            }

Como último paso ejecutamos el spider en el terminal:     'scrapy crawl imdb_spider -o movies.json'
